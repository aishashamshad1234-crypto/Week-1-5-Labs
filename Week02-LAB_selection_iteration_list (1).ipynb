12. **Dog Years**

It is commonly said that one human year is equivalent to 7 dog years. However, this simple conversion fails to recognise that dogs reach adulthood in approximately two years. As result, some people believe that it is better to count each of the first two human years as 10.5 dog years, and then count each additional human year as 4 dog years. 

Write a program that implements the conversion from human years to dog years described in the previous paragraph. Ensure that your program works correctly for conversions of less than two or more humans’ years. Your program should display an appropriate error message if the user enters a negative number. 


````python

# Ask the user to enter human years
human_years = float(input("Enter human years: "))

# Check if the input is negative
if human_years < 0:
    print("Error: Years cannot be negative.")
# If human years are 2 or less, multiply by 10.5 for dog years
elif human_years <= 2:
    dog_years = human_years * 10.5
# If more than 2 years, first two years = 21 dog years, remaining years = 4 dog years each
else:
    dog_years = 21 + (human_years - 2) * 4

# Display the result
print(f"Equivalent dog years: {dog_years}")
````

15. **Find the highest score**

Write a program that prompts the user to enter the number of students and each student’s name and score and display the student name and highest score.


````python
# Program: Find the highest score and the student's name

# Ask the user how many students there are
n = int(input("Enter number of students: "))

# Check for invalid number of students
if n <= 0:
    print("Error: Number of students must be a positive integer.")
else:
    # Start with the lowest possible score and an empty name
    highest_score = float('-inf')
    highest_name = ""

    # Read each student's name and score
    for i in range(1, n + 1):
        name = input(f"Enter name of student {i}: ")
        score = float(input(f"Enter {name}'s score: "))

        # If this score is higher than the current highest, update both
        if score > highest_score:
            highest_score = score
            highest_name = name
    # Show the result (if multiple have the same top score, this shows the first one entered)
    print(f"Highest score: {highest_score} by {highest_name}")
````

16. Write a **guessing game** in Python where the user has to guess a secret number. After every guess the program tells the user whether their number was too large or too small. At the end the number of tries needed should be printed.


````python
# Guessing Game: The computer picks a secret number, you try to guess it.

import random  # To generate a random number

secret = random.randint(1, 100)  # Secret number between 1 and 100
tries = 0  # Counts how many guesses the user makes

while True:
    # Ask the user to guess and convert the input to an integer
    guess = int(input("Guess the number (between 1 and 100): "))
    tries += 1  # Increase the try count

    # Tell the user if the guess is too small, too large, or correct
    if guess < secret:
        print("Too small!")
    elif guess > secret:
        print("Too large!")
    else:
        print(f"Correct! You guessed it in {tries} tries.")
        break  # Exit the loop when the correct number is guessed
````

17. **Collecting Positive Numbers**
You are writing a small program to help a teacher keep track of positive numbers entered by students. The teacher also wants to know if any even numbers were given.

**Requirements:**

1. Keep asking the user to enter a positive number, or type `"done"` to stop.
2. Store all entered numbers in a **list**.
3. After the input ends:

   * If numbers were entered:

     * Print all the numbers entered.
     * Check the list and print the **first even number**. If no even number exists, print `"No even numbers found"`.
   * If no numbers were entered, print `"No numbers were entered"`.


> ### Note 
>
>* **`lower()`**: Converts any text into lowercase.
>  Example: `"DONE".lower()` becomes `"done"`.
>  This makes it easier to compare user input without worrying about uppercase/lowercase.
>
>* **`isdigit()`**: Checks if a string only contains digits (0–9).
>  Example: `"123".isdigit()` returns `True`, but `"12a".isdigit()` returns `False`.
>  This helps us make sure the user typed a valid number before converting it into an integer.
> 


---

````python
# Program: Collecting Positive Numbers and Finding the First Even Number

numbers = []  # List to store all valid positive numbers

while True:
    # Ask the user for input (number or 'done' to finish)
    user_input = input('Enter a positive number (or type "done" to finish): ').strip()

    # Convert the input to lowercase to compare safely (e.g., 'DONE' -> 'done')
    if user_input.lower() == "done":
        break  # Exit the loop when the user is done

    # Check if the input is made of digits only (rules out negatives, decimals, and letters)
    if not user_input.isdigit():
        print("Invalid input. Please enter digits only (e.g., 12) or 'done'.")
        continue  # Ask again

    # Convert the valid digit string to an integer
    num = int(user_input)

    # Ensure the number is positive (> 0)
    if num <= 0:
        print("Please enter a positive number greater than 0.")
        continue

    # Add the valid positive number to the list
    numbers.append(num)

# After input ends

# If no numbers were entered
if not numbers:
    print("No numbers were entered")
else:
    # Print all the numbers entered
    print("Numbers entered:", numbers)

    # Find the first even number using a simple loop
    first_even = None
    for n in numbers:
        if n % 2 == 0:   # Check if the number is even
            first_even = n
            break        # Stop at the first even number

    # Print the result based on whether an even number was found
    if first_even is None:
        print("No even numbers found")
    else:
        print("First even number:", first_even)
````

18. Suppose you want to develop a **program to play a lottery.**
    
The program randomly generates a two-digit number, prompts the user to enter a <u>two-digit number</u>, and determines whether the user wins according to the following rules:


    1. If the user’s input matches the lottery in the exact order, the award is £10,000.
    2. If all the digits in the user’s input match all the digits in the lottery number, the award is £3,000.
    3. If one digit in the user’s input matches a digit in the lottery number, the award is £1,000.



````python
# Lottery Game: Match a two-digit number

import random  # To generate a random two-digit number

# Generate a random two-digit lottery number as a *string* with leading zero if needed (e.g., "07")
lottery = f"{random.randint(0, 99):02d}"

# Ask the user to enter a two-digit number
user = input("Enter a two-digit number (00-99): ").strip()

# Validate input: must be exactly two digits (allows "00" to "99")
if not (user.isdigit() and len(user) == 2):
    print("Error: Please enter exactly two digits (e.g., 07, 23, 99).")
else:
    # Show the lottery number
    print(f"Lottery number: {lottery}")

    # Check the winning rules in order of highest prize to lowest
    if user == lottery:
        # Exact match (same digits in the same order)
        print("Exact match! You win £10,000.")
    elif sorted(user) == sorted(lottery):
        # All digits match but order is different (e.g., user=21, lottery=12)
        print("All digits match (wrong order). You win £3,000.")
    elif set(user) & set(lottery):
        # At least one digit matches between the two numbers
        print("One digit matches. You win £1,000.")
    else:
        # No match at all
        print("No match. Better luck next time!")
````

    1. If the user input matches the lottery number in the exact order, the award is £10,000.
    2. If all the digits in the user input match all the digits in the lottery number, the award is £3,000.

````python
# Lottery Game (Three-Digit): Match a number and win prizes

import random  # To generate a random three-digit number

# Generate a random lottery number between 100 and 999 (no leading zeros)
lottery = random.randint(100, 999)
lottery_str = str(lottery)  # Keep a string version for easy digit comparisons

# Ask the user to enter a valid three-digit number (100–999)
user = input("Enter a three-digit number (100-999): ").strip()

# Validate input: must be digits and within the 100–999 range
if not (user.isdigit() and 100 <= int(user) <= 999):
    print("Error: Please enter a valid three-digit number between 100 and 999.")
else:
    # Reveal the lottery number
    print(f"Lottery number: {lottery_str}")

    # 1) Exact match (same digits in the same order)
    if user == lottery_str:
        print("Exact match! You win £10,000.")
    # 2) All digits match but order is different
    elif sorted(user) == sorted(lottery_str):
        print("All digits match (wrong order). You win £3,000.")
    # 3) At least one digit matches anywhere
    elif set(user) & set(lottery_str):
        print("One digit matches. You win £1,000.")
    # No match at all
    else:
        print("No match. Better luck next time!")
````

20. (Find the two highest scores) - Write a program that prompts the user to enter the number of students and each student’s score and displays the highest and second highest scores.


````python
# Program: Find the highest and second highest scores

# Ask how many students there are
n = int(input("Enter number of students: "))

# Validate number of students
if n <= 0:
    print("Error: Number of students must be a positive integer.")
else:
    scores = []  # List to store all scores

    # Read each student's score
    for i in range(1, n + 1):
        score = float(input(f"Enter score for student {i}: "))
        scores.append(score)

    # Remove duplicates to get distinct scores, then sort descending
    distinct_scores = sorted(set(scores), reverse=True)

    # We need at least two distinct scores to have a "second highest"
    if len(distinct_scores) < 2:
        # Either only one student or all scores are the same
        print("Highest score:", distinct_scores[0])
        print("No second highest score (not enough distinct scores).")
    else:
        highest = distinct_scores[0]
        second_highest = distinct_scores[1]
        print("Highest score:", highest)
        print("Second highest score:", second_highest)
````


21. **Password cracker** - write a program to print the accuracy of password guess.

For example: The actual password is `MDX123` while a user tried `MDX321` or `mdx123`, the program should tell that how accurately the password was guessed, in this case it should be `50%` accuracy. if a user enter `MDX124`, the output should be `83%` and in case of `123MDX` the output should be `0%`.

Points to keep in mind while calculating the accuracy a) case of the letters entered b) order of the letters

If a password is accurately guessed, the output should state `Password cracked!`

````python
# Program: Password Accuracy Checker

# Read the actual password and the user's guess
actual = input("Enter the actual password: ")
guess  = input("Enter your guess: ")

# If lengths differ, it's not a fair comparison (since order and positions matter)
if len(actual) != len(guess):
    print("Error: Password and guess must have the same length.")
else:
    # If it's an exact match (same chars, same case), it's cracked
    if guess == actual:
        print("Password cracked!")
    else:
        # Collect indices for letters and digits based on the ACTUAL password
        letter_indices = []
        digit_indices  = []

        for i, ch in enumerate(actual):
            if ch.isalpha():
                letter_indices.append(i)
            elif ch.isdigit():
                digit_indices.append(i)
            # (If there are other characters, they won't be scored in this simple version)

        # Score letters (exact char & case on the letter positions) 
        letter_matches = 0
        if len(letter_indices) > 0:
            for i in letter_indices:
                if guess[i] == actual[i]:   # case-sensitive match at the same position
                    letter_matches += 1
            letter_pct = (letter_matches / len(letter_indices)) * 100
        else:
            letter_pct = None  # No letters to score

        # Score digits (exact digit on the digit positions)
        digit_matches = 0
        if len(digit_indices) > 0:
            for i in digit_indices:
                if guess[i] == actual[i]:   # exact digit at the same position
                    digit_matches += 1
            digit_pct = (digit_matches / len(digit_indices)) * 100
        else:
            digit_pct = None  # No digits to score

        # Combine scores 
        # If we have both letters and digits, average them equally (fits your examples).
        # If we only have one type, use that score. If neither (rare), fall back to per-char accuracy.
        if letter_pct is not None and digit_pct is not None:
            accuracy = round((letter_pct + digit_pct) / 2)
        elif letter_pct is not None:
            accuracy = round(letter_pct)
        elif digit_pct is not None:
            accuracy = round(digit_pct)
        else:
            # Fallback: per-character exact matches (handles special-character passwords)
            exact_matches = sum(1 for i in range(len(actual)) if actual[i] == guess[i])
            accuracy = round((exact_matches / len(actual)) * 100)

        print(f"Accuracy: {accuracy}%")
````
