4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‘@’.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def validate_password():
    password = input("Enter your password: ")

    if ' ' in password:
        print("Error: Password must not contain spaces.")
        return False

    valid = True
    msg =""

    if not any(ch.isupper() for ch in password):
        msg = msg + ("\nError: Must contain at least one uppercase letter.")
        valid = False
    if not any(ch.islower() for ch in password):
        msg = msg +("\nError: Must contain at least one lowercase letter.")
        valid = False
    if not any(ch.isdigit() for ch in password):
        msg = msg +("\nError: Must contain at least one digit.")
        valid = False
    if not any(ch in "!@#$%^&*()-_=+[]{};:'\",.<>/?\\|`~" for ch in password):
        msg = msg +("\nError: Must contain at least one special character.")
        valid = False

    if valid :
        print("Password is valid!")
    else:
        print("Password is invalid." + msg)

validate_password()
````

9. (Optional)
Basic Regex for Pattern Matching
Import the re module and use it to search for an email address in a string.
Use regex to find all the words that start with a capital letter in a sentence.

````python
# Basic Regex for Pattern Matching

import re  # regex module

# Task 1: Email search
# A simple email pattern: name@domain.tld
email_pattern = r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}"

text = input("Enter some text that may contain an email address: ").strip()

# Find all email addresses in the text
emails = re.findall(email_pattern, text)

if emails:
    print("First email found:", emails[0])
    print("All emails found:", emails)
else:
    print("No email address found.")

print("-" * 50)

# Task 2: Capitalized words
# Matches words starting with a capital letter (A–Z), followed by letters
# This will match 'London', 'Alice', 'UAE' (all caps), etc.
capitalized_pattern = r"\b[A-Z][a-zA-Z]*\b"

sentence = input("Enter a sentence: ").strip()

# Find all words that start with a capital letter
capitalized_words = re.findall(capitalized_pattern, sentence)

if capitalized_words:
    print("Words starting with a capital letter:", capitalized_words)
else:
    print("No capitalized words found.")
````

````python
# Function to count lowercase, uppercase, digits, and special symbols
def count_char_types(text):
    # Initialize counters
    lower = 0
    upper = 0
    digits = 0
    special = 0

    # Go through each character and update the right counter
    for ch in text:
        if ch.islower():        # a-z
            lower += 1
        elif ch.isupper():      # A-Z
            upper += 1
        elif ch.isdigit():      # 0-9
            digits += 1
        else:                   # Anything else (spaces, punctuation, symbols, etc.)
            special += 1

    # Print the results
    print("Lowercase letters:", lower)
    print("Uppercase letters:", upper)
    print("Digits:", digits)
    print("Special symbols:", special)

    # (Optional) also return the counts as a dict if you want to use them later
    return {
        "lower": lower,
        "upper": upper,
        "digits": digits,
        "special": special
    }
print("Enter a password to analyze: " )
password = input()
count_char_types(password)

````

13. Secret Word Generator with Randomization: Ask the user for their first name and surname. Create a function that generates a special secret word that:
Uses the first 3 letters of the first name and last 2 letters of the surname.
Randomly appends a special character (e.g., @, #, $, etc.) at the end of the secret word.

```python
special_chars = "@#$%^&*!"
```

````python
import random

# Function to generate a secret word
def generate_secret_word(first_name, surname):
    # Take first 3 letters of first name and last 2 letters of surname
    part1 = first_name[:3]      # first 3 letters
    part2 = surname[-2:]        # last 2 letters

    # Special characters list
    special_chars = "@#$%^&*!"

    # Randomly pick one special character
    random_char = random.choice(special_chars)

    # Combine parts to form the secret word
    secret_word = part1 + part2 + random_char
    return secret_word

# ---- Example usage ----
first_name = input("Enter your first name: ")
surname = input("Enter your surname: ")

secret = generate_secret_word(first_name, surname)
print("Your secret word is:", secret)
````

15. (Binary to decimal) Write a function that parses a binary number as a string and convert it into a decimal integer. Use the function header: def binaryToDecimal(binaryString): For example, binary string 10001 is 17 (1 24 + 0
23 + 0 22 + 0 21 + 1 * 20 = 17)

So, binaryToDecimal("10001") returns 17.

Write a test program that prompts the user to enter a binary string and displays the corresponding decimal integer value.

````python
# Function: Convert a binary string to a decimal integer (without using built-ins)
def binaryToDecimal(binaryString):

    total = 0

    for ch in binaryString:
        # Validate each character (must be '0' or '1')
        if ch not in ('0', '1'):
            raise ValueError("Invalid binary string: only '0' and '1' are allowed.")
        # Shift left by 1 (multiply by 2) and add current bit
        total = total * 2 + int(ch)

    return total


# Ask the user for a binary string and display the decimal value
binary_input = input("Enter a binary string (e.g., 10001): ").strip()

# Basic validation before calling the function
if binary_input == "":
    print("Error: Input cannot be empty.")
elif not all(c in "01" for c in binary_input):
    print("Error: Please enter only 0s and 1s.")
else:
    decimal_value = binaryToDecimal(binary_input)
    print(f"Binary: {binary_input} -> Decimal: {decimal_value}")
````

16. (Binary to hex) Write a function that parses a binary number and converts itinto a hex number. The function header is: def binaryToHex(binaryValue):
Write a test program that prompts the user to enter a binary number and displays the corresponding hexadecimal value.



````python
# Function: Convert a binary string to a hexadecimal string (uppercase)
def binaryToHex(binaryValue):
    """
    Converts a binary string (e.g., "10001") to its hexadecimal representation (e.g., "11").
    - Accepts an optional '0b'/'0B' prefix.
    - Raises ValueError for invalid input (non-binary characters or empty string).
    """
    if not isinstance(binaryValue, str):
        raise ValueError("Input must be a string.")

    s = binaryValue.strip()

    # Allow optional binary prefix like '0b1010'
    if s.startswith(("0b", "0B")):
        s = s[2:]

    # Basic validation: must be non-empty and contain only 0 or 1
    if s == "" or not all(ch in "01" for ch in s):
        raise ValueError("Invalid binary number: use only 0 and 1 (e.g., 101101).")

    # Convert: binary -> int -> hex string (strip '0x' and make uppercase)
    return hex(int(s, 2))[2:].upper()


# Prompt the user and display the corresponding hexadecimal value
user_input = input("Enter a binary number (e.g., 10001 or 0b1011): ")

try:
    hex_value = binaryToHex(user_input)
    print(f"Binary: {user_input.strip()} -> Hex: {hex_value}")
except ValueError as err:
    print("Error:", err)
````

17. (Decimal to binary) Write a function that converts a decimal number into a binary number as a string. Use the function header: def decimalToBinary(value):
Write a test program that prompts the user to enter a decimal integer value and displays the corresponding binary value.

````python
# Function to convert a decimal number to a binary string
def decimalToBinary(value):
    # If the number is 0, just return "0"
    if value == 0:
        return "0"
    
    binary = ""  # This will store the binary result
    
    # Keep dividing the number by 2 and store the remainder
    while value > 0:
        remainder = value % 2          # Get remainder (0 or 1)
        binary = str(remainder) + binary  # Add it to the front of the binary string
        value = value // 2             # Divide the number by 2 (integer division)
    
    return binary  # Return the final binary string

# Ask the user to enter a decimal number
decimal_value = int(input("Enter a decimal number: "))

# Call the function to convert it to binary
binary_value = decimalToBinary(decimal_value)

# Display the binary result
print("The binary value is:", binary_value)
````
