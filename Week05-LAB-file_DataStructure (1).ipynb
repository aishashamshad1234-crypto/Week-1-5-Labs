### Activity 3: Temperature Logger

Write a program that:
1. Asks the user for temperatures for 5 days
2. Writes each temperature to "temps.txt" with the day number
   Format: "Day 1: 22°C"
3. Reads the file and calculates the average temperature

<hr>

`````python
# Activity 3: Ask 5 temps, write "Day X: YY°C" to file, read back and average

temps = []

# Collect temperatures for 5 days
for day in range(1, 6):
    while True:
        t = input(f"Enter temperature for Day {day} (e.g., 22): ").strip()
        if t.lstrip("-").isdigit():      # allow negative integers
            temps.append(int(t))
            break
        else:
            print("Please enter a valid integer temperature.")

# Write to file in the required format
with open("temps.txt", "w", encoding="utf-8") as f:
    for day, t in enumerate(temps, start=1):
        f.write(f"Day {day}: {t}°C\n")

# Read file and calculate average
read_temps = []
with open("temps.txt", "r", encoding="utf-8") as f:
    for line in f:
        # Line format: "Day X: YY°C" -> extract YY
        try:
            after_colon = line.split(":", 1)[1].strip()  # "YY°C"
            number_str = after_colon.split("°", 1)[0]    # "YY"
            read_temps.append(int(number_str))
        except Exception:
            pass  # skip malformed lines

if read_temps:
    avg = sum(read_temps) / len(read_temps)
    print("Temperatures read:", read_temps)
    print(f"Average temperature: {avg:.2f}°C")
else:
    print("No temperatures found in file.")
`````

### **Activity 5: Unique Hobbies Set** 


Two friends share their hobbies:

Find and print:

- Hobbies only Alex enjoys
- All unique hobbies combined

---

`````python
# Activity 5: Set operations with hobbies

alex_hobbies = {"reading", "gaming", "cooking", "hiking"}
sam_hobbies  = {"gaming", "hiking", "photography", "music"}

# Hobbies both enjoy (intersection)
both = alex_hobbies & sam_hobbies

# Hobbies only Alex enjoys (difference)
only_alex = alex_hobbies - sam_hobbies

# All unique hobbies combined (union)
all_unique = alex_hobbies | sam_hobbies

print("Hobbies both enjoy:", both)
print("Hobbies only Alex enjoys:", only_alex)
print("All unique hobbies combined:", all_unique)
`````

### Activity 9: Finding "Dream" Sentences in Les_Brown.txt

From the `Les_Brown.txt` file, read the content and identify all sentences that contain the word "dream". Write these sentences to a new file called `dream_sentences.txt`.

**Steps:**
1. Open and read the content of `Les_Brown.txt`.
2. Split the text into sentences.
3. Filter out sentences that contain the word "dream" (case-insensitive).
4. Write these sentences into the new file `dream_sentences.txt`.
<hr>

`````python
import re

def extract_dream_sentences(input_file, output_file):
    # Step 1: Read the content of Les_Brown.txt
    with open(input_file, "r", encoding="utf-8") as f:
        text = f.read()

    # Step 2: Split text into sentences
    # This regex splits on '.', '?', or '!' but keeps the delimiter.
    sentences = re.split(r'(?<=[.!?])\s+', text)

    # Step 3: Filter sentences containing "dream" (case-insensitive)
    dream_sentences = [
        sentence for sentence in sentences
        if "dream" in sentence.lower()
    ]

    # Step 4: Write these sentences into dream_sentences.txt
    with open(output_file, "w", encoding="utf-8") as f:
        for sentence in dream_sentences:
            f.write(sentence.strip() + "\n")

# Example usage:
extract_dream_sentences("Les_Brown.txt", "dream_sentences.txt")
`````

### Activity 11: Adding Line Numbers to Python Zen

Create a Python program that reads the content from the `python_zen.txt` file and adds line numbers to each line. Each line in the output file should start with a line number, followed by a dot and a space, and then the original line from the input file. Skip numbering the first line as it is the title. Save the result into a new file called `numbered_zen.txt`.

**Steps:**
1. Open the `python_zen.txt` file for reading.
2. Skip the first line (which is the title) when numbering.
3. For each subsequent line, prepend the line number followed by a dot and a space.
4. Write the modified content to a new file called `numbered_zen.txt`.

<hr>

`````python
def number_zen_lines(input_file, output_file):
    # Step 1: Open the python_zen.txt file for reading
    with open(input_file, "r", encoding="utf-8") as f:
        lines = f.readlines()

    # Prepare a list to store the modified lines
    numbered_lines = []

    # Step 2 & 3: Skip the first line (title) and number the rest
    for i, line in enumerate(lines):
        if i == 0:
            # Keep the title unchanged
            numbered_lines.append(line)
        else:
            # Add line numbers starting from 1
            numbered_lines.append(f"{i}. {line}")

    # Step 4: Write the modified content to numbered_zen.txt
    with open(output_file, "w", encoding="utf-8") as f:
        f.writelines(numbered_lines)


`````

### Activity 12: Reversing Clothing Stock List

Create a program that reads the `clothing_stock.txt` file and stores all the clothing items in a list. Then reverse the order of the items and write the reversed list into a new file called `reversed_clothing_stock.txt`.

**Steps:**
1. Read all clothing items into a list.
2. Reverse the list.
3. Write the reversed list into a new file.
<hr>

`````python
def reverse_clothing_stock(input_file, output_file):
    # Step 1: Read all clothing items into a list
    with open(input_file, "r", encoding="utf-8") as f:
        items = f.readlines()

    # Remove newline characters for cleaner output
    items = [item.rstrip("\n") for item in items]

    # Step 2: Reverse the list
    reversed_items = list(reversed(items))

    # Step 3: Write the reversed list into a new file
    with open(output_file, "w", encoding="utf-8") as f:
        for item in reversed_items:
            f.write(item + "\n")
`````

### Activity 13: Identifying Duplicate Entries in Clothing Stock

Create a program that reads the `clothing_stock.txt` file and identifies if there are any duplicate entries (both item and size). Use a set to find and print these duplicates.

**Steps:**
1. Read the file and store each item in a set.
2. Keep track of duplicate entries.
3. Print any duplicates found.
<hr>

`````python
def find_duplicates(input_file):
    seen = set()        # Stores unique entries
    duplicates = set()  # Stores duplicates found

    # Step 1: Read the file and store each item in a set
    with open(input_file, "r", encoding="utf-8") as f:
        for line in f:
            entry = line.strip()

            if entry in seen:
                # Step 2: Track duplicates
                duplicates.add(entry)
            else:
                seen.add(entry)

    # Step 3: Print duplicates
    if duplicates:
        print("Duplicate entries found:")
        for item in duplicates:
            print(item)
    else:
        print("No duplicates found.")
`````

### Activity 14: Counting Different Sizes per Clothing Item

Create a Python program that reads data from the `clothing_stock.txt` file and counts how many different sizes of each clothing item are available. Store this in a dictionary and print it.

**Steps:**
1. Read the file and extract both clothing items and sizes.
2. Count how many different sizes exist for each item.
3. Store the results in a dictionary and print it.
<hr>

`````python

def count_sizes(input_file):
    size_counts = {}

    # Step 1: Read the file and extract clothing items and sizes
    with open(input_file, "r", encoding="utf-8") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue  # skip empty lines
            
            # Split into item and size
            item, size = line.split(",")

            item = item.strip()
            size = size.strip()

            # Step 2: Count how many different sizes exist for each item
            if item not in size_counts:
                size_counts[item] = set()
            size_counts[item].add(size)

    # Convert sets to counts
    final_counts = {item: len(sizes) for item, sizes in size_counts.items()}

    # Step 3: Print the dictionary
    print(final_counts)
`````

### Activity 15: Managing Study Groups

You are responsible for managing a study group for a course. Each student can be part of multiple study groups, and you need to track which groups a student is enrolled in. Students can join or leave study groups. What data structure would you choose to manage this efficiently, and why?

**Example data:**

```python
{"John": ["Math", "Physics"], "Emily": ["Physics", "Biology"], "Alex": ["Math"]}

Steps:

1. Choose a data structure that allows quick access to the groups each student belongs to.
2. Write a program that lets students join and leave study groups.
3. Implement a function to display all groups a student is part of.
<hr>

`````python
study_groups = {
    "John": {"Math", "Physics"},
    "Emily": {"Physics", "Biology"},
    "Alex": {"Math"}
}

# Function to add a student to a group
def join_group(student, group):
    if student not in study_groups:
        study_groups[student] = set()
    study_groups[student].add(group)
    print(f"{student} has joined {group}.")

# Function to remove a student from a group
def leave_group(student, group):
    if student in study_groups:
        study_groups[student].discard(group)
        print(f"{student} has left {group}.")
    else:
        print(f"{student} is not in the system.")

# Function to display all groups of a student
def show_groups(student):
    if student in study_groups:
        groups = study_groups[student]
        if groups:
            print(f"{student} is enrolled in: {', '.join(groups)}")
        else:
            print(f"{student} is not currently enrolled in any groups.")
    else:
        print(f"{student} is not in the system.")
`````
